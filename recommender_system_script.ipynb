```python
import ast
```

```python
def fetch_name(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
len(movies)
```

```python
movies_with_full_data = movies.dropna()
len(movies_with_full_data)
```

```python
movies_without_full_data = movies[~movies['movie_id'].isin(movies_with_full_data['movie_id'])]
movies_without_full_data #overview,keywords and cast not available.
```

```python
cols = ['genres', 'keywords']
for col in cols:
    movies_with_full_data[col] = movies_with_full_data[col].apply(fetch_name)

movies_with_full_data.head()
```

```python
# movies_with_full_data['keywords'] = movies_with_full_data['keywords'].apply(fetch_name)
# movies_with_full_data.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def fetch_first_3_cast_names(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies_with_full_data['cast'] = movies_with_full_data['cast'].apply(fetch_first_3_cast_names)
movies_with_full_data.head()
```

```python
movies_with_full_data['cast'] = movies_with_full_data['cast'].apply(lambda x:x[0:3])
movies_with_full_data.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies_with_full_data['crew'] = movies_with_full_data['crew'].apply(fetch_director)
movies_with_full_data.head()
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies_with_full_data.head().sample(5)
```

```python
movies_with_full_data.crew.tolist()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python

cols= ['cast','crew','genres','keywords']

for col in cols:
    movies_with_full_data[col] = movies_with_full_data[col].apply(collapse)

movies_with_full_data.head()
# movies_with_full_data['crew'] = movies_with_full_data['crew'].apply(collapse)
# movies_with_full_data['genres'] = movies_with_full_data['genres'].apply(collapse)
# movies_with_full_data['keywords'] = movies_with_full_data['keywords'].apply(collapse)
```

```python
movies_with_full_data['overview'] = movies_with_full_data['overview'].apply(lambda x:x.split())
```

```python
movies_with_full_data['tags'] = movies_with_full_data['overview'] + movies_with_full_data['genres'] + movies_with_full_data['keywords'] + movies_with_full_data['cast'] + movies_with_full_data['crew']
```

```python
new = movies_with_full_data.drop(columns=['overview','genres','keywords','cast','crew'])
# new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
    
```

```python
recommend('Ironman')
```
